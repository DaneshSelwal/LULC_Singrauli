```python
def plot_best_scores(best_scores, excel_file_path, class_labels):
    # Extract the best pruner for each model based on Accuracy and F1-score
    best_acc_scores = {}
    best_f1_scores = {}

    for (model_name, pruner_name), scores in best_scores.items():
        if model_name not in best_acc_scores:
            best_acc_scores[model_name] = (scores['test_accuracy'], pruner_name)
        if model_name not in best_f1_scores:
            f1 = scores['test_f1'] if scores['test_f1'] is not None else 0
            best_f1_scores[model_name] = (f1, pruner_name)

        if scores['test_accuracy'] > best_acc_scores[model_name][0]:
            best_acc_scores[model_name] = (scores['test_accuracy'], pruner_name)
        f1 = scores['test_f1'] if scores['test_f1'] is not None else 0
        if f1 > best_f1_scores[model_name][0]:
            best_f1_scores[model_name] = (f1, pruner_name)

    # Prepare data for plotting Accuracy
    model_names_acc = [f"{model} ({pruner})" for model, (acc, pruner) in best_acc_scores.items()]
    acc_values = [acc for acc, _ in best_acc_scores.values()]

    # Prepare data for plotting F1
    model_names_f1 = [f"{model} ({pruner})" for model, (f1, pruner) in best_f1_scores.items()]
    f1_values = [f1 for f1, _ in best_f1_scores.values()]

    # Plot Accuracy
    plt.figure(figsize=(12, 6))
    bars_acc = plt.bar(model_names_acc, acc_values, color='skyblue')
    best_acc_index = np.argmax(acc_values)
    bars_acc[best_acc_index].set_color('orange')
    for i, bar in enumerate(bars_acc):
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 0.05,
                 f'{acc_values[i]:.5f}', ha='center', va='bottom', color='black')
    plt.xlabel('Model (Pruner)')
    plt.ylabel('Test Accuracy')
    plt.title('Best Test Accuracy for Each Model')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    acc_image_path = 'accuracy_plot.png'
    plt.savefig(acc_image_path)
    plt.close()

    # Plot F1 score
    plt.figure(figsize=(12, 6))
    bars_f1 = plt.bar(model_names_f1, f1_values, color='lightgreen')
    best_f1_index = np.argmax(f1_values)
    bars_f1[best_f1_index].set_color('orange')
    for i, bar in enumerate(bars_f1):
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 0.05,
                 f'{f1_values[i]:.5f}', ha='center', va='bottom', color='black')
    plt.xlabel('Model (Pruner)')
    plt.ylabel('Test F1-score')
    plt.title('Best Test F1-score for Each Model')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    f1_image_path = 'f1_plot.png'
    plt.savefig(f1_image_path)
    plt.close()

    # (OPTIONAL) Plot ROC-AUC, if desired and present in your scores (uncomment if needed)
    best_auc_scores = {}
    for (model_name, pruner_name), scores in best_scores.items():
        auc = scores['test_auc'] if scores['test_auc'] is not None else 0
        if model_name not in best_auc_scores or auc > best_auc_scores[model_name][0]:
            best_auc_scores[model_name] = (auc, pruner_name)
    model_names_auc = [f"{model} ({pruner})" for model, (auc, pruner) in best_auc_scores.items()]
    auc_values = [auc for auc, _ in best_auc_scores.values()]
    if auc_values:
        plt.figure(figsize=(12, 6))
        bars_auc = plt.bar(model_names_auc, auc_values, color='tomato')
        best_auc_index = np.argmax(auc_values)
        bars_auc[best_auc_index].set_color('orange')
        for i, bar in enumerate(bars_auc):
            plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 0.05,
                     f'{auc_values[i]:.5f}', ha='center', va='bottom', color='black')
        plt.xlabel('Model (Pruner)')
        plt.ylabel('Test ROC-AUC')
        plt.title('Best Test ROC-AUC for Each Model')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        auc_image_path = 'auc_plot.png'
        plt.savefig(auc_image_path)
        plt.close()

    # Load the existing Excel file
    workbook = load_workbook(excel_file_path)
    sheet_name = 'Best Models Plots'
    if sheet_name in workbook.sheetnames:
        sheet = workbook[sheet_name]
    else:
        sheet = workbook.create_sheet(title=sheet_name)

    # Insert images into the new Excel sheet
    img_acc = Image(acc_image_path)
    img_f1 = Image(f1_image_path)
    sheet.add_image(img_acc, 'A1')
    sheet.add_image(img_f1, 'A20')  # Adjust as needed

    # If using AUC plot
    # if auc_values:
    #     img_auc = Image(auc_image_path)
    #     sheet.add_image(img_auc, 'A39')

    workbook.save(excel_file_path)

    # Clean up the image files
    os.remove(acc_image_path)
    os.remove(f1_image_path)
    # if auc_values:
    #     os.remove(auc_image_path)

# Example class labels
CLASS_LABELS = {
    0: 'Forest',
    1: 'Water',
    2: 'Agriculture & Vegetation',
    3: 'Settlement',
    4: 'Mining',
    5: 'Barren Land',
}
```

# **Autosampler by Optuna**

```python
optuna.logging.set_verbosity(optuna.logging.WARNING)

def hyperparameter_tuning_all(X_train, y_train, X_test, y_test):
    X_train = np.array(X_train)
    y_train = np.array(y_train)
    X_test = np.array(X_test)
    y_test = np.array(y_test)

    models = {
        'Random Forest': (RandomForestClassifier, {
            'n_estimators': [100, 200, 300, 500],
            'criterion': ['gini', 'entropy', 'log_loss'],
            'max_depth': [None, 10, 20, 30, 40],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 3, 5],
            'max_features': ['sqrt', 'log2', None],
            'max_leaf_nodes': [None, 50, 100, 200],
            'n_jobs': [-1],
            'random_state': [42],
            'verbose': [0],
            'class_weight': [None, 'balanced']
        }),
        'Gradient Boosting': (GradientBoostingClassifier, {
            'loss': ['log_loss'],
            'learning_rate': [0.01, 0.05, 0.1, 0.2],
            'n_estimators': [100, 200, 300],
            'subsample': [1.0, 0.9, 0.7],
            'criterion': ['friedman_mse', 'squared_error'],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 3, 5],
            'max_depth': [3, 5, 7, 10],
            'random_state': [42],
            'max_features': [None, 'sqrt', 'log2'],
            'verbose': [0],
            'max_leaf_nodes': [None, 10, 30],
            'warm_start': [False]
        }),
        'XGBoost': (XGBClassifier, {
            'n_estimators': [100, 200, 300],
            'learning_rate': [0.01, 0.05, 0.1, 0.15],
            'max_depth': [3, 5, 7],
            'min_child_weight': [1, 3, 5],
            'gamma': [0, 0.1, 0.5, 1],
            'subsample': [0.5, 0.7, 0.9],
            'colsample_bytree': [0.5, 0.7, 0.9],
            'reg_alpha': [0, 0.01, 0.1, 1],
            'reg_lambda': [0.1, 1, 5, 10],
            'objective': ['binary:logistic'],  # For multiclass: 'multi:softprob'
            'random_state': [42],
            'n_jobs': [-1],
            'use_label_encoder': [False],
            'eval_metric': ['logloss'],
        }),
        'LightGBM': (LGBMClassifier, {
            'n_estimators': [100, 200, 300],
            'learning_rate': [0.01, 0.05, 0.1, 0.15],
            'num_leaves': [15, 31, 63],
            'max_depth': [3, 5, 7, -1],
            'min_child_samples': [1, 5, 10, 20],
            'subsample': [0.5, 0.7, 0.9, 1.0],
            'colsample_bytree': [0.5, 0.7, 0.9, 1.0],
            'reg_alpha': [0, 0.01, 0.1, 1],
            'reg_lambda': [0, 0.1, 1, 10],
            'min_child_weight': [1e-5, 1e-3, 1e-2],
            'random_state': [42],
            'n_jobs': [-1],
            'verbose': [-1]
        }),
        'CatBoost': (CatBoostClassifier, {
            'iterations': [200, 500],
            'learning_rate': [0.01, 0.03, 0.05, 0.1],
            'depth': [4, 6, 8, 10],
            'l2_leaf_reg': [1, 3, 5, 7, 9],
            'border_count': [32, 64, 128],
            'random_seed': [42],
            'verbose': [0]
        }),
        # 'HistGradientBoosting': (HistGradientBoostingClassifier, {
        #     'learning_rate': [0.01, 0.05, 0.1],
        #     'max_iter': [100, 200, 300],
        #     'max_depth': [3, 5, 7, None],
        #     'min_samples_leaf': [5, 10, 20],
        #     'max_leaf_nodes': [15, 31, 63, None],
        #     'l2_regularization': [0.0, 0.1, 0.5],
        #     'random_state': [42]
        # }),
    }

    pruners = [
        optuna.pruners.MedianPruner(),
        optuna.pruners.NopPruner(),
        optuna.pruners.PatientPruner(optuna.pruners.MedianPruner(), patience=3),
        optuna.pruners.PercentilePruner(25.0),
        optuna.pruners.SuccessiveHalvingPruner(),
        optuna.pruners.HyperbandPruner(),
        optuna.pruners.ThresholdPruner(lower=0.1),
        optuna.pruners.WilcoxonPruner()
    ]

    best_scores = {}

    for model_name, (model_class, param_space) in models.items():
        for pruner in pruners:
            pruner_name = pruner.__class__.__name__
            print(f"Running Optuna for {model_name} with {pruner_name}...")
            try:
                study = optuna.create_study(direction='maximize', pruner=pruner)

                def objective(trial, model_class=model_class, param_space=param_space):
                    params = {}
                    for key, values in param_space.items():
                        if isinstance(values, list):
                            params[key] = trial.suggest_categorical(key, values)
                        elif isinstance(values, tuple):
                            if len(values) == 2:
                                params[key] = trial.suggest_float(key, values[0], values[1])
                            elif len(values) == 3 and isinstance(values[2], bool) and values[2]:
                                params[key] = trial.suggest_int(key, values[0], values[1])
                            else:
                                raise ValueError(f"Invalid parameter range for {key}")
                    model = model_class(**params)
                    model.fit(X_train, y_train)
                    y_pred = model.predict(X_test)
                    acc = accuracy_score(y_test, y_pred)
                    return acc

                study.optimize(objective, n_trials=1)
                best_params = study.best_params
                best_model = model_class(**best_params)
                best_model.fit(X_train, y_train)
                y_pred = best_model.predict(X_test)

                acc = accuracy_score(y_test, y_pred)
                f1 = None
                auc = None
                try:
                    f1 = f1_score(y_test, y_pred, average='macro')
                except:
                    pass
                try:
                    if hasattr(best_model, "predict_proba"):
                        if len(np.unique(y_test)) == 2:
                            auc = roc_auc_score(y_test, best_model.predict_proba(X_test)[:, 1])
                        else:
                            auc = roc_auc_score(y_test, best_model.predict_proba(X_test), multi_class='ovr')
                except:
                    pass

                best_scores[(model_name, pruner_name)] = {
                    'best_score': acc,
                    'best_params': best_params,
                    'test_accuracy': acc,
                    'test_f1': f1,
                    'test_auc': auc,
                    'pruner': pruner_name
                }
                print(f"Best ACC for {model_name} with {pruner_name}: {acc} with params: {best_params}")
                print(f"Best F1 for {model_name} with {pruner_name}: {f1}")
                print(f"Best ROC AUC for {model_name} with {pruner_name}: {auc}")
            except Exception as e:
                print(f"Failed to run Optuna for {model_name} with {pruner_name}. Error: {e}")

    if best_scores:
        best_model_name, best_pruner_name = max(best_scores, key=lambda k: best_scores[k]['test_accuracy'])
        best_model_info = best_scores[(best_model_name, best_pruner_name)]
        print(f"\nBest model on test data: {best_model_name} with {best_pruner_name}")
        print(f"Test Accuracy: {best_model_info['test_accuracy']}")
        print(f"Test F1: {best_model_info['test_f1']}")
        print(f"Test ROC AUC: {best_model_info['test_auc']}")
        print(f"Best Parameters: {best_model_info['best_params']}")
        print(f"Pruner Used: {best_model_info['pruner']}")
    else:
        print("No valid model configurations found.")

    return best_scores

best_scores_autosampler = hyperparameter_tuning_all(X_train, y_train, X_test, y_test)
```

# **Best Model Analysis**

```python
# Update for classification model classes
model_mapping = {
    'Random Forest': RandomForestClassifier,
    'Gradient Boosting': GradientBoostingClassifier,
    'XGBoost': XGBClassifier,
    'LightGBM': LGBMClassifier,
    # 'GPBoost': GPBoostClassifier,   # uncomment if available
    'CatBoost': CatBoostClassifier,
    # 'NGBoost': NGBClassifier,       # uncomment if used
    # 'HistGradientBoosting': HistGradientBoostingClassifier, # optional
    # 'PGBM': PGBMClassifier         # if available
}

def get_best_models_and_predict(best_scores_autosampler, X_train, y_train, X_test, y_test, file_path, class_labels):
    import numpy as np
    import pandas as pd
    from sklearn.metrics import confusion_matrix, classification_report
    import matplotlib.pyplot as plt
    import seaborn as sns
    import os

    # Convert input data to NumPy arrays
    X_train = np.array(X_train)
    y_train = np.array(y_train)
    X_test = np.array(X_test)
    y_test = np.array(y_test)

    # Dictionary to store the best model for each type
    best_models = {}

    # Iterate over the dictionary to find the best pruner for each model type
    for (model_name, pruner), params in best_scores_autosampler.items():
        current_score = params.get('test_accuracy', -np.inf)  # maximize accuracy!
        if model_name not in best_models or current_score > best_models[model_name]['score']:
            best_models[model_name] = {
                'score': current_score,
                'params': params['best_params'],
                'pruner': pruner
            }

    # Prepare a DataFrame to store predictions
    df = pd.read_csv(file_path)

    # Iterate over the best models to train and predict
    for model_name, model_info in best_models.items():
        best_params = model_info['params']
        model_class = model_mapping.get(model_name)

        if model_class is None:
            print(f"Model {model_name} is not supported or not available.")
            continue

        # Handle specific parameters or settings for CatBoost
        if model_name == 'CatBoost':
            best_params.pop('verbose', None)

        # Create an instance of the best model with the best parameters
        model = model_class(**best_params)
        model.fit(X_train, y_train)
        predictions = model.predict(X_test)

        # If predictions are probabilities, convert to class indices
        if isinstance(predictions, np.ndarray) and predictions.ndim > 1:
            predictions = np.argmax(predictions, axis=1)

        # Map numerical predictions to class names
        class_name_predictions = [class_labels.get(pred, 'Unknown') for pred in predictions.flatten()]

        # Add predictions to the DataFrame using class names
        df[f'{model_name} Predictions'] = class_name_predictions

        # Plot confusion matrix for classification
        cm = confusion_matrix(y_test, predictions)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=list(class_labels.values()),
                    yticklabels=list(class_labels.values()))
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title(f'Confusion Matrix ({model_name})')
        plt.tight_layout()

        # Save the plot temporarily
        plot_path = f'temp_plot_{model_name}.png'
        plt.savefig(plot_path)
        plt.close()

        # Optionally print the classification report using class names
        print(f"Classification Report for {model_name}:")
        print(classification_report(y_test, predictions, target_names=list(class_labels.values())))

    # Save predictions and plots to Excel
    with pd.ExcelWriter(file_path.replace('.csv', '_results.xlsx'), engine='xlsxwriter') as writer:
        # Write data to Excel
        df.to_excel(writer, sheet_name='Data', index=False)

        # Get the xlsxwriter objects
        workbook = writer.book

        # Insert each plot into a separate worksheet
        for model_name in best_models.keys():
            # Shorten the worksheet name to fit within Excel's 31 character limit
            short_model_name = ''.join([word[0] for word in model_name.split()])  # e.g., 'RF' for 'Random Forest'
            sheet_name = f'{short_model_name}_Plot'

            worksheet = workbook.add_worksheet(sheet_name)
            writer.sheets[sheet_name] = worksheet
            plot_path = f'temp_plot_{model_name}.png'
            worksheet.insert_image('A1', plot_path)

    # Clean up temporary plot files
    for model_name in best_models.keys():
        os.remove(f'temp_plot_{model_name}.png')

    return df, best_models


# Call the function with your class labels
CLASS_LABELS = {
    0: 'Forest',
    1: 'Water',
    2: 'Agriculture & Vegetation',
    3: 'Settlement',
    4: 'Mining',
    5: 'Barren Land',
}

df, best_models = get_best_models_and_predict(best_scores_autosampler, X_train, y_train, X_test, y_test, "./drive/MyDrive/Singrauli_uncertainity/test.csv", CLASS_LABELS)
```
